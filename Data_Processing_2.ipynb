```python
import pandas as pd
df = pd.read_csv('five_cuisine_df.csv')
```

```python
# rename column 'stars_y' to 'stars'

df.rename(columns={'stars_y': 'stars'}, inplace=True)
```

```python
# In stars column, change the value of 1 to 3 to string name 'negative', and change the value of 4 and 5 to string name 'positive'

df['stars'] = df['stars'].replace([1,2,3], 'negative')
df['stars'] = df['stars'].replace([4,5], 'positive')
```

```python
drop_columns = ['city', 'state', 'postal_code', 'stars_x', 'review_count', 'attributes', 'user_id', 'review_id', 'useful', 'funny', 'cool']
df = df.drop(drop_columns, axis=1)
```

```python
df.head()
```

```python
mexican_df = df[df['category']=='Mexican']
american_df = df[df['category']=='American']
japanese_df = df[df['category']=='Japanese']
chinese_df = df[df['category']=='Chinese']
italian_df = df[df['category']=='Italian']
```
